```python
#Dropping columns
```

```python
df2=df.drop(["arrival_date_week_number","adr","company","agent"],axis=1)
```

```python
df2.isnull().sum()
```

```python
df2["children"]=df["children"].fillna(df2["children"].mode()[0])
df2["country"]=df["country"].fillna(df2["country"].mode()[0])
```

```python
df2.isnull().sum()
```

```python
df2.children = df2.children.astype('int64')
```

```python
#for reservation status no show and canceled "is_canceled" status is 1.So we can drop these two columns to avoid repetition of data
```

```python
df2['arrival_date'] =  df2['arrival_date_day_of_month'].astype('str')+df2['arrival_date_month'].astype('str')+df2['arrival_date_year'].astype('str')  
```

```python
df2['arrival_date'] = pd.to_datetime(df2['arrival_date'])
df2['reservation_status_date'] = pd.to_datetime(df2['reservation_status_date'])
```

```python
df2['days_diff'] = (df2['reservation_status_date'] - df2['arrival_date'])
df2.head()
```

```python
df3=df2.drop(['arrival_date_year','arrival_date_month','arrival_date_day_of_month','reservation_status','reservation_status_date'],axis=1)
```

```python
df3['day_of_week'] = df3.arrival_date.dt.dayofweek  # Monday is 0 and Sunday is 6
```

```python
df3['is_weekend'] = (df3['day_of_week'] >=5).astype('int')
df3.head()
```

```python
# Outliers treatment
```

```python
df3['lead_time'].plot(kind='box')
plt.show()
```

```python
Q1 = df3['lead_time'].quantile(0.25)
Q3 = df3['lead_time'].quantile(0.75)
IQR = Q3 - Q1

LL = Q1 - 1.5 * IQR
UL = Q3 + 1.5 * IQR

print(f"Q1: {Q1} | Q3: {Q3} | IQR: {IQR} | LL: {LL} | UL: {UL}")

outlier_count_ul = df3[df3['lead_time'] > UL].shape[0]
outlier_perc_ul = (outlier_count_ul * 100) / df3.shape[0]

outlier_count_ll = df3[df3['lead_time'] < LL].shape[0]
outlier_perc_ll = (outlier_count_ll * 100) / df3.shape[0]

print(f"# of Outliers on the Higher Side: {outlier_count_ul} | % of Outliers on the Higher Side: {outlier_perc_ul}%")
print(f"# of Outliers on the Lower Side: {outlier_count_ll} | % of Outliers on the Lower Side: {outlier_perc_ll}%")
```

```python
# Creating new variable with Z-score of each record:
# 'SciPy' is used to perform scientific computations
import scipy.stats as stats
df3['ZR'] = stats.zscore(df3['lead_time'])
```

```python
df3.head()
```

```python
### Cleaned Data: without outliers so z>-3 and z< +3
df4= df3[(df3['ZR']>-3) & (df3['ZR']<3)].reset_index()
df4.head()
```

```python
# count of outliers:

df3[(df3['ZR']<-3) | (df3['ZR']>3)].shape[0]
```

```python
df3.shape
```

```python
df4.shape
```

```python
#Interpretation:
#A crude way to know whether the outliers have been removed or not is to check the dimensions of the data. From the above output, we can see that the dimensions are reduced that implies outliers are removed.
```

# MACHINE LEARNING


```python
# Features and Target

X = df6.drop('is_canceled',axis=1)
Y = df6['is_canceled']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
#Decision Tree model
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(xtrain,ytrain)
```

```python
ytrainpred = dt_model.predict(xtrain)
ytestpred = dt_model.predict(xtest)
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(ytrain,ytrainpred))
```

```python
print(metrics.classification_report(ytest,ytestpred))
```
